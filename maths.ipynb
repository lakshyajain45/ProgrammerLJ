```python
# write a program to evaluate
#1!+2!+3!+4!+5!...+n!
import math
n=int(input("enter any number: "))
sum=0
for i in range(1,n+1):
  fact=math.factorial(i)
  sum+=fact
print(sum)
```

```python
# write a program to enter the number till the user wants and display the maximum number entered using while loop
max_num = int(input('Enter a number (or type "stop" to finish): '))
while True:
    user_input = input('Enter another number (or type "stop" to finish): ')
    if user_input.lower() == 'stop':
        break
    try:
        num = int(user_input)
        if num > max_num:
            max_num = num
    except ValueError:
        print("Invalid input. Please enter a number or 'stop'.")
print(f"The maximum number entered is: {max_num}")
```

```python
class car:
  def __init__(self,brand,model,year):#self is a reference, calling object, adress
       self.brand=brand #variables
       self.model=model
       self.year=year

  def display(self):
    print(f'car: {self.brand} {self.model} {self.year} ')

car1=car('toyota','corolla',2022)
car2=car('tesla','model s',2023)
car1.display()
car2.display()
```

```python
class student:
  def __init__(self,name,age,grade):
    self.name=name
    self.age=age
    self.grade=grade

  def display(self):
    print(f'name: {self.name} age: {self.age} grade: {self.grade}')
#list
student=[
    student('john',18,'A'),
    student('jane',19,'B')
]
for s in student:
  s.display()
```
