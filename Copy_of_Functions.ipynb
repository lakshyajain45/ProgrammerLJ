```python
# Demonstrates local versus global variable scope. `x` is initialized globally to 10. Inside `funct1`, a local `x` is created, multiplied by 10, and printed. After the function call, the global `x` is printed, showing it remains unchanged.
x=10
def funct1():
  x=10
  x= x*10
  print(x)

funct1()
print(x)
```

```python
# Demonstrates how to access and modify a global variable from within a function using `globals()`. `x` is initialized globally to 10. Inside `funct1`, the global `x` is accessed, multiplied by 10, and updated globally. Both prints show the modified global `x`.
x=10
def funct1():
  x=globals()['x']
  x=10
  x= x*10
  globals()['x']=x
  print(x)

funct1()
print(x)
```

```python
# Demonstrates how to modify a global variable from within a function using the `global` keyword. `x` is initialized globally to 10. Inside `funct1`, the global `x` is directly modified. Both prints show the updated global `x`.
x=10
def funct1():
  global x
  x= x*50
  print(x)

funct1()
print(x)
```

```python
#  Function to check if a number is even or odd
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Demonstration
print(f"The number 7 is: {check_even_odd(7)}")
```

```python
#  Function to calculate the area of a circle
import math
def circle_area(radius):
    return math.pi * (radius ** 2)

# Demonstration
print(f"Area of a circle with radius 5 is: {circle_area(5):.2f}")
```

```python
#  Function to reverse a string
def reverse_string(text):
    return text[::-1]

# Demonstration
print(f"Reversed string: {reverse_string('Hello World')}")
```

```python
# Function to check if a string is a palindrome
def is_palindrome(text):
    clean_text = text.replace(" ", "").lower()
    return clean_text == clean_text[::-1]

# Demonstration
word = "Radar"
print(f"Is '{word}' a palindrome? {is_palindrome(word)}")
```

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Demonstration
temp_c = 25
print(f"{temp_c}°C is equal to {celsius_to_fahrenheit(temp_c)}°F")
```

```python
# Function to calculate the factorial of a number
def calculate_factorial(n):
    if n < 0:
        return "Error: Factorial not defined for negative numbers"
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Demonstration
num = 5
print(f"The factorial of {num} is: {calculate_factorial(num)}")
```

```python
# Function to generate the Fibonacci sequence up to n terms
def fibonacci_sequence(n):
    sequence = [0, 1]
    while len(sequence) < n:
        sequence.append(sequence[-1] + sequence[-2])
    return sequence[:n]

# Demonstration
terms = 10
print(f"The first {terms} terms of Fibonacci: {fibonacci_sequence(terms)}")
```

```python
# Function simulating a simple calculator
def simple_calculator(a, b, operation):
    if operation == '+':
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '*':
        return a * b
    elif operation == '/':
        return a / b if b != 0 else "Division by zero error"
    else:
        return "Invalid Operation"

# Demonstration
print(f"Addition (10 + 5): {simple_calculator(10, 5, '+')}")
print(f"Multiplication (10 * 5): {simple_calculator(10, 5, '*')}")
```
